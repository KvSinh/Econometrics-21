```python
import pandas as pd
import numpy as np                      #Importing Libraries
import matplotlib.pyplot as plt
```

```python
df1= pd.read_excel('wage.xlsx', sheet_name= 'Sheet1', engine='openpyxl')        #Importing DataFrame
```

```python
df1
```

```python
df1.drop('year', axis=1, inplace= True)
df1.drop('maritl',axis=1, inplace= True)
df1.drop('race', axis=1, inplace= True)
df1.drop('region', axis=1, inplace= True)
df1.drop('health', axis=1, inplace= True)
df1.drop('health_ins', axis=1, inplace= True)
df1.drop('wage', axis=1, inplace= True)                      #Dropping not useful variables
```

```python
name= df1['education'].str.split('.')
```

```python
second=[]
for i in range (0,3000,1):
  second.append(name[i][1])                         #data cleaning, removing 1,2,3,4 from education
```

```python
df1["education"]= second
```

```python
df1
```

```python
name2= df1['jobclass'].str.split('.')
```

```python
temp=[]
for i in range (0,3000,1):                         #datacleaning, removing 1. ,2.,3.,4. from jobclass
  temp.append(name2[i][1])
```

```python
df1['jobclass']= temp
```

```python
df1.columns = [x.strip() for x in df1.columns]
df1
```

```python
df1= pd.get_dummies(df1, prefix=['educ','jobclass'], drop_first=True)  #Introducing Dummies
df1
```

```python
Y= df1['logwage']                           #Defining X and Y variables
X= df1.drop('logwage',axis=1)                          
```

```python
Y
```

```python
import statsmodels.api as sm
X_new=sm.add_constant(X)                                 #Running Regression of Logwage on age, education and Jobclass
model = sm.OLS(Y, X_new)
results = model.fit()
print(results.summary())
```

```python
X
```

```python
age= X['age']
jobclass= X['jobclass_ Information']
jobclass_new=sm.add_constant(jobclass)                      #Running Regression of age on jobclass
model1 = sm.OLS(age, jobclass_new)
results1 = model1.fit()
```

```python
age_star=results1.resid
age_star                          
```

```python
educ1= X['educ_ Advanced Degree']
educ2= X['educ_ College Grad']
educ3= X['educ_ HS Grad']
educ4= X['educ_ Some College']
                                                                  #Running Regression of education on jobclass
model2 = sm.OLS(educ1, jobclass_new)
results2 = model2.fit()
educ1_star=results2.resid

model3 = sm.OLS(educ2, jobclass_new)
results3 = model3.fit()
educ2_star=results3.resid

model4 = sm.OLS(educ3, jobclass_new)
results4 = model4.fit()
educ3_star=results4.resid

model5 = sm.OLS(educ4, jobclass_new)
results5 = model5.fit()
educ4_star=results5.resid
```

```python
final=  { 'agestar': age_star, 'educ1star': educ1_star, 'educ2star': educ2_star, 'educ3star': educ3_star, 'educ4star': educ4_star }
```

```python
Xf= pd.DataFrame(final)                             #Defining a dataframe to run regression in next step
```

```python
Xfinal=sm.add_constant(Xf)
model_final = sm.OLS(Y, Xfinal)                     #Final Regression
results_final = model_final.fit()
print(results_final.summary())
```
